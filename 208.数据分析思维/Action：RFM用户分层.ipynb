# 用户分层 RFM

```python
data["order date (DateOrders)"] = pd.to_datetime(data["order date (DateOrders)"])
# 统计最后一笔订单的时间
data["order date (DateOrders)"].max()
```

```python
# 假设当前时间为2018-2-1
import datetime
now = datetime.datetime(2018,2,1)

# 计算每个用户的RFM指标
customer_seg = data.groupby('Customer Id').agg({'order date (DateOrders)': \
            lambda x: (now-x.max()).days, 'Customer Id':lambda x:len(x), 'Sales':lambda x: x.sum()})
```

```python
customer_seg
```

```python
customer_seg.rename(columns={'order date (DateOrders)':'R_value', 'Customer Id': 'F_value', 'Sales':'M_value'}, inplace = True)

```

```python
customer_seg
```

```python
# 将RFM数据划分为2个尺度
quantiles = customer_seg.quantile(q=[0.5])
```

```python
quantiles = quantiles.to_dict()
quantiles
```

```python
def R_Score(x,b,c):
    if x<=c[b][0.5]:
        return 2
    else:
        return 1

def FM_Score(x,b,c):
    if x>=c[b][0.5]:
        return 2
    else:
        return 1
```

```python
# 创建R_Score,用于将R_value => [1,2]
customer_seg['R_Score'] = customer_seg['R_value'].apply(R_Score,args=('R_value',quantiles))
```

```python
# 创建F_Score,用于将F_value => [1,2]
customer_seg['F_Score'] = customer_seg['F_value'].apply(FM_Score, args = ('F_value', quantiles))
# 创建M_Score,用于将M_value => [1,2]
customer_seg['M_Score'] = customer_seg['M_value'].apply(FM_Score, args = ('M_value', quantiles))
customer_seg
```

```python
# 计算RFM用户分层
def RFM_user(df):
    if df['R_Score'] == 2 and df['F_Score'] == 2 and df['M_Score'] ==2:
        return '重要价值用户'
    elif df['R_Score'] == 2 and df['F_Score'] == 1 and df['M_Score'] ==2:
        return '重要发展用户'
    elif df['R_Score'] == 1 and df['F_Score'] == 2 and df['M_Score'] ==2:
        return '重要保持用户'
    elif df['R_Score'] == 1 and df['F_Score'] == 1 and df['M_Score'] ==2:
        return '重要挽留用户'
    elif df['R_Score'] == 2 and df['F_Score'] == 2 and df['M_Score'] ==1:
        return '一般价值用户'
    elif df['R_Score'] == 2 and df['F_Score'] == 1 and df['M_Score'] ==1:
        return '一般发展用户'
    elif df['R_Score'] == 1 and df['F_Score'] == 2 and df['M_Score'] ==1:
        return '一般保持用户'
    elif df['R_Score'] == 1 and df['F_Score'] == 1 and df['M_Score'] ==1:
        return '一般挽留用户'
```

```python
customer_seg['Customer_Type'] = customer_seg.apply(RFM_user, axis=1)
customer_seg
```
