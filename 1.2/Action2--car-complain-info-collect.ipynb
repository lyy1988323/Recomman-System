```python
from lxml import etree
selector=etree.HTML(response_text)
```

```python
ret_tittle = selector.xpath('//*[@id="content"]/div[4]/div[2]/table/tr/th/text()')
ret_tittle
```

```python
# 获取到大部分信息：'投诉编号', '投诉品牌', '投诉车系', '投诉车型', '典型问题', '投诉时间'
ret_content = selector.xpath('//*[@id="content"]/div[4]/div[2]/table/tr/td/text()')
ret_content
```

```python
# 获取‘问题简述’
ret_content1 = selector.xpath('//*[@id="content"]/div[4]/div[2]/table/tr/td/a/text()')
ret_content1
```

```python
# 获取‘典型问题’
ret_content2 = selector.xpath('//*[@id="content"]/div[4]/div[2]/table/tr/td/em/text()')
ret_content2
```

```python
ret_content_list1 = np.array(ret_content).reshape(-1,6).tolist()
ret_content_list1
```

```python
# 结果按顺序拼接
ret = []
for index,item in enumerate(ret_content_list1):
#     item = list(item)
    item.insert(4,ret_content1[index])
    item.append(ret_content2[index])
    ret.append(item)
ret
```

```python
# 结果写入csv存档
ret_pd=pd.DataFrame(columns=ret_tittle,data=ret)
ret_pd.set_index(["投诉编号"], inplace=True)
print(ret_pd)
ret_pd.to_csv('complain_infos.csv',encoding='utf-8')
```
