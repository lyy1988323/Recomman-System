```python
import pandas as pd
import numpy as np
```

```python
# 导入数据
origin_data = pd.read_csv("Market_Basket/Market_Basket_Optimisation.csv",header=None)
```

```python
origin_data.head()
```

```python
origin_data.shape
```

```python
# 处理缺失值
data = origin_data.fillna(0)
data.head()
```

```python
# 创建购物篮
transactions = []
for row in range(origin_data.shape[0]):
    temp_set = set()
    for item in data.iloc[row,:]:
        if item:
            temp_set.add(item)
    transactions.append(temp_set)
transactions[:5]
```

```python
itemsets, rules = apriori(transactions, min_support=0.05,  min_confidence=0.1)
```

```python
from mlxtend.frequent_patterns import apriori as apriori_
from mlxtend.frequent_patterns import association_rules
```

```python
# 导入数据
sr=pd.read_csv("Market_Basket/Market_Basket_Optimisation.csv",header=None,sep='\t')
sr.head()
```

```python
one_hot_data = sr[0].str.get_dummies(sep=',')
one_hot_data.head()
```

```python
frequent_itemsets = apriori_(one_hot_data, min_support=0.05, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=0.1)
```

```python
print("频繁项集：", frequent_itemsets)
print("关联规则：", rules)
```
